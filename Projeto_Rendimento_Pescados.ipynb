```python
import pandas as pd
import win32com.client as win32
import plotly.express as px
import os
import kaleido
import numpy as np
```

```python
# SALMAO

# importa a base de dados

teste = pd.read_excel(r'C:\Users\rodrigo.souza\Desktop\PROJETO RENDIMENTO\pastateste.xlsx', sheet_name='1.05S')

# visualiza a base

pd.set_option('display.max_columns',None)  # com essa linha de código nós estamos fazendo o pandas mostrar todas as colunas dessa tabela, e o NONE significa que não queremos limite, ou seja, se quisessemos um limite era só colocar um número.

#print(teste)
#print('-' * 50)

# visualizar somente categoria, data, período, loja, peso e ano

tabela = teste[['CATEGORIA','DATA','PERIODO','LOJA','PESO (KG)','ANO']]

#print(tabela)
#print('-'*50)

rendimento = tabela.loc[tabela['CATEGORIA']=='RENDIMENTO']

#print(rendimento)
#print('-'*50)

ano = rendimento.loc[rendimento['ANO']==2022]

#print(ano)
#print('-'*50)

periodo = ano.loc[ano['PERIODO']=='julho'].groupby('DATA').mean()


#print(periodo)
#print('-'*50)

mensal = ano.loc[ano['PERIODO']=='julho'].groupby('PERIODO').mean()

#print(mensal)
#print('-'*50)

#CRIANDO O GRÁFICO
grafico = px.bar(periodo, y='PESO (KG)', text_auto= True)
grafico.write_image('rendimentosalmao.pdf', engine="kaleido")

#MOSTRANDO O GRÁFICO
#grafico.show()

#----------------------------------------------------------------------------------------------

# ATUM

# importa a base de dados

teste1 = pd.read_excel(r'C:\Users\rodrigo.souza\Desktop\PROJETO RENDIMENTO\pastateste.xlsx', sheet_name='1.05A')

# visualiza a base

pd.set_option('display.max_columns',None)  # com essa linha de código nós estamos fazendo o pandas mostrar todas as colunas dessa tabela, e o NONE significa que não queremos limite, ou seja, se quisessemos um limite era só colocar um número.

#print(teste1)
#print('-' * 50)

# visualizar somente categoria, data, período, loja, peso e ano

tabela1 = teste1[['CATEGORIA','DATA','PERIODO','LOJA','PESO (KG)','ANO']]

#print(tabela1)
#print('-'*50)

rendimento1 = tabela1.loc[tabela1['CATEGORIA']=='RENDIMENTO']

#print(rendimento1)
#print('-'*50)

ano1 = rendimento1.loc[rendimento1['ANO']==2022]

#print(ano1)
#print('-'*50)

periodo1 = ano1.loc[ano1['PERIODO']=='julho'].groupby('DATA').mean()


#print(periodo1)
#print('-'*50)

mensal1 = ano1.loc[ano1['PERIODO']=='julho'].groupby('PERIODO').mean()

#print(mensal1)
#print('-'*50)

#CRIANDO O GRÁFICO
grafico1 = px.bar(periodo1, y='PESO (KG)', text_auto= True)
grafico1.write_image('rendimentoatum.pdf', engine="kaleido")

#MOSTRANDO O GRÁFICO
#grafico1.show()

print('tudo ok')
```
