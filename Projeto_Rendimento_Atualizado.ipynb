```python
import pandas as pd
import win32com.client as win32
import plotly.express as px
import os
import kaleido
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SALMAO

# importa a base de dados

teste = pd.read_excel('pastateste.xlsx', sheet_name='1.05S')
teste['DATA'] = pd.to_datetime(teste['DATA'], format='%d/%m/%Y', errors='coerce')
# visualiza a base

pd.set_option('display.max_columns',None)  # com essa linha de código nós estamos fazendo o pandas mostrar todas as colunas dessa tabela, e o NONE significa que não queremos limite, ou seja, se quisessemos um limite era só colocar um número.

#print(teste)
#print('-' * 50)

# visualizar somente categoria, data, período, loja, peso e ano

tabela = teste[['CATEGORIA','DATA','PERIODO','LOJA','PESO (KG)','ANO']] # aqui nós estamos criando uma nova tabela, chamada tabela, e estamos selecionando somente as colunas de categoria, data, período, loja, peso e ano, para facilitar a visualização dos dados.

#print(tabela)
#print('-'*50)

rendimento = tabela.loc[(tabela['CATEGORIA']=='RENDIMENTO') & 
                        (tabela['LOJA']=='RIO SUL') & 
                        (tabela['PERIODO']=='julho')] # aqui nós estamos filtrando os dados para mostrar somente a categoria de rendimento, e como só temos uma categoria, então o resultado vai ser a média do peso para cada data do mês de julho.

#print(rendimento)
#print('-'*50)

periodo = rendimento.groupby('DATA')['PESO (KG)'].mean()

#print(periodo)
#print('-'*50)

mensal = rendimento.groupby('PERIODO')['PESO (KG)'].mean()

#print(mensal)
#print('-'*50)

funcionario = teste.loc[(teste['ANO']==2022) & 
                (teste['CATEGORIA']=='RENDIMENTO') & 
                (teste['PERIODO']=='julho') & 
                (teste['LOJA']=='RIO SUL')].groupby('FUNCIONARIO')['PESO (KG)'].mean()


#CRIANDO O GRÁFICO
plt.figure(figsize=(12,6))
grafico = plt.bar(periodo.index, height=(periodo*100) ,color='black', alpha=0.7, edgecolor='green',
                  align='center', width=0.8, ) # o height=(periodo*100) é para multiplicar os valores do período por 100, transformando-os em porcentagem, e o color='black' é para definir a cor das barras como preta, o alpha=0.7 é para definir a transparência das barras, o edgecolor='green' é para definir a cor da borda das barras como verde, e o align='center' é para centralizar as barras no eixo x.
plt.title('Rendimento Salmão - Julho/2022', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Rendimento (%)', fontsize=12)
plt.axhline(y=67, color='red', linestyle='--', linewidth=2, label='Média Esperada - 67%') # adiciona uma linha horizontal no gráfico, e o y=67 é para definir a posição da linha no eixo y, o color='red' é para definir a cor da linha como vermelha, o linestyle='--' é para definir o estilo da linha como tracejada, o linewidth=2 é para definir a espessura da linha, e o label='Média Esperada' é para adicionar um rótulo à linha, que será exibido na legenda do gráfico.
plt.bar_label(grafico, fmt='%.2f%%', padding=-55, color='white', fontsize=12,
              rotation=90, label_type='edge') # adiciona os rótulos de valor em cada barra, e o fmt='%.2f%%' é para formatar os valores como porcentagem com duas casas decimais, o padding=-16 é para ajustar a posição dos rótulos dentro das barras, o color='white' é para definir a cor dos rótulos como branco, e o fontsize=12 é para definir o tamanho da fonte dos rótulos.
plt.xticks(rotation=45)
plt.legend(loc='lower right', fontsize=10, shadow=True) # adiciona a legenda no gráfico, e o loc='lower right' é para posicionar a legenda no canto inferior direito do gráfico, e o fontsize=10 é para definir o tamanho da fonte da legenda, e o shadow=True é para adicionar uma sombra à legenda, deixando-a mais destacada no gráfico.

plt.savefig('rendimentosalmao.pdf', format='pdf', bbox_inches='tight') # salva o gráfico em formato pdf, e o bbox_inches='tight' é para ajustar o layout do gráfico, evitando que partes importantes sejam cortadas.

plt.close() # fecha o gráfico para não mostrar na tela, mas salva o arquivo em pdf. Além disso, fechando o gráfico o próximo não irá sobrepor o gráfico anterior.
```
